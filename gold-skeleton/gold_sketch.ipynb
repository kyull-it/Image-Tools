```python
# !pip install --upgrade imutils
# !pip install --upgrade cv2

import cv2
import imutils
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
imagePath = r'test_skeleton.png'
imagePath2 = r'test_image.jpg'
```

```python
logo = Image.open(imagePath)  #numpy array로 불러와지는지 확인?
#window로 사진 보여줌
logo.show()
logo
```

```python
plt.imshow(logo)
# plot형태로 보여줌
plt.show()
```

```python
logo = cv2.imread(imagePath)
# logo
```

```python
# skeletonize the image
gray = cv2.cvtColor(logo, cv2.COLOR_BGR2GRAY)  #무조건 gray로 변환 필요
skeleton = imutils.skeletonize(gray, size=(3, 3))
cv2.imshow("Skeleton", skeleton)
#jupyter에서만 필요한 함수 (에러방지)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## image RGB color extract
gold_color = cv2.imread('gold_color.png', cv2.IMREAD_UNCHANGED)
r,g,b = gold_color[1,1]

gold_skeleton = cv2.imread('skeleton_Bulgook-sa.jpg')
# gold_skeleton = cv2.cvtColor(gray_skeleton, cv2.COLOR_GRAY2RGB) 
gold_skeleton[np.where((gold_skeleton>[20, 20, 20]).all(axis=2))] = [r, g, b]
cv2.imwrite('gold_skeleton_Bulgook-sa.jpg', gold_skeleton)
cv2.imshow("gold_skeleton", gold_skeleton)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
