```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# 전자기 스펙트럼 범위 정의
spectrum_ranges = {
    "Radio waves": (1e3, 1e-1),
    "Microwaves": (1e-1, 1e-3),
    "Infrared": (1e-3, 7e-7),
    "Visible light": (7e-7, 4e-7),
    "Ultraviolet": (4e-7, 1e-8),
    "X-rays": (1e-8, 1e-11),
    "Gamma rays": (1e-11, 1e-14)
}
```

```python

# 스펙트럼 범위 시각화
fig, ax = plt.subplots(figsize=(14, 2))
colors = plt.cm.plasma(np.linspace(0, 1, len(spectrum_ranges)))

for (label, (start, end)), color in zip(spectrum_ranges.items(), colors):
    ax.plot([start, end], [1, 1], linewidth=20, color=color, label=label)

ax.set_xscale('log')
ax.set_yticks([])
ax.set_xlabel('Wavelength (meters)')
ax.legend(loc='upper center', ncol=7, bbox_to_anchor=(0.5, 1.5))
ax.set_title('Electromagnetic Spectrum')

plt.show()
```
